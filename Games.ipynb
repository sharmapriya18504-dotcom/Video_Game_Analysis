```python
import pandas as pd
```

```python
df = pd.read_csv("games.csv")
df
```

```python
df.head()
```

```python
df.info()
```

```python
df.drop(columns=["Unnamed: 0"], inplace=True)
```

```python
df.isna().sum()
```

```python
## Rating
df["Rating"] = pd.to_numeric(df["Rating"], errors="coerce")
df["Rating"]= df["Rating"].fillna(df["Rating"].mean())

## Team
df["Team"] = df["Team"].fillna("Unknown")

## Summary
df["Summary"]= df["Summary"].fillna("No summary")
```

```python
df.isnull().sum()
```

```python
def convert_k(val):
    if isinstance(val, str):
        val = val.replace(",", "").strip()
        if "K" in val:
            return float(val.replace("K", "")) * 1000
        try:
            return float(val)
        except:
            return None
    return val

cols = ["Times Listed", "Number of Reviews", "Plays", "Playing", "Backlogs", "Wishlist"]

for col in cols:
    df[col] = df[col].apply(convert_k)
```

```python
def convert_k(x):
    if isinstance(x, str) and 'K' in x:
        return float(x.replace('K','')) * 1000
    return x

cols = ["plays","playing","backlogs","wishlist","times_listed","number_of_reviews"]

for col in cols:
    if col in df.columns:
        df[col] = df[col].apply(convert_k)
```

```python
import re

def clean_text(val):
    if isinstance(val, str):
        # remove quotes and brackets
        val = re.sub(r"[\'\"\[\]]", "", val)
        return val.strip()
    return val

text_cols = ["Title", "Team", "Genres", "Summary", "Reviews"]

for col in text_cols:
    df[col] = df[col].apply(clean_text)
```

```python
# 1. column names clean
df.columns = df.columns.str.lower().str.replace(" ", "_")

# 2. special characters remove
df = df.replace(r'[^\x00-\x7F]+', '', regex=True)

# 3. quotes + brackets remove
df = df.replace({'"': '', "'": '', r'\[': '', r'\]': ''}, regex=True)
```
